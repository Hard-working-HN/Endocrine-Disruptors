#### Mouse(F) Carcinogen Score

```python
Xtrain_0,Xtest_0,Ytrain_0,Ytest_0 = train_test_split(X, Y_0, test_size=0.3, random_state=472)
params = {
    'booster': 'gbtree',
    'objective': 'reg:squarederror',
    'gamma': 0,
    'max_depth': 2,
    'lambda': 4,
    'subsample': 1,
    'colsample_bytree': 1,
    'min_child_weight': 1,
    'verbosity': 1,
    'eta': 0.19,
    'random_state': 420,
    'nthread': 4,
}
dtrain_0 = xgb.DMatrix(Xtrain_0, Ytrain_0)
num_rounds = 800
plst = list(params.items())
model = xgb.train(plst, dtrain_0, num_rounds)
dtest_0 = xgb.DMatrix(Xtest_0, Ytest_0)
ans = model.predict(dtest_0)
labels = dtest_0.get_label()
mse = mean_squared_error(labels, ans)
rmse = mean_squared_error(labels, ans, squared=False)  # RMSE
r2 = r2_score(labels, ans)
mape = mean_absolute_percentage_error(labels, ans)
print(f"Mean Absolute Percentage Error (MAPE): {mape:.8f}")
print("Mouse(F) Carcinogen Score:")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")

importance = model.get_score(importance_type='gain')
sorted_importance = sorted(importance.items(), key=operator.itemgetter(1), reverse=True)
print(pd.DataFrame(sorted_importance))
Z = pd.DataFrame(sorted_importance)
Z.to_excel(r"Mouse(F) Carcinogen Score.xlsx")
```
